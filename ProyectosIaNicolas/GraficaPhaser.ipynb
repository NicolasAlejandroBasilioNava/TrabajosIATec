```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Cargar los datos desde el CSV
df = pd.read_csv('./phaserData.csv', header=None, names=['x1', 'y1', 'x2', 'y2', 't1', 't2'], sep=' ', dtype=float)

# Crear la figura
fig = plt.figure()

# Agregar un subplot 3D
ax = fig.add_subplot(111, projection='3d')

# Graficar con ts en 0
sc = ax.scatter(df[df['t1'] == 0]['x1'], df[df['t1'] == 0]['y1'], df[df['t1'] == 0]['t1'], c='gray', marker='o', label='Jump = 0')
sc = ax.scatter(df[df['t2'] == 0]['x2'], df[df['t2'] == 0]['y2'], df[df['t2'] == 0]['t2'], c='brown', marker='o', label='Move = 0')

# Graficar con ts en 1
sc = ax.scatter(df[df['t1'] == 1]['x1'], df[df['t1'] == 1]['y1'], df[df['t1'] == 1]['t1'], c='orange', marker='o', label='Jump = 1')
sc = ax.scatter(df[df['t2'] == 1]['x2'], df[df['t2'] == 1]['y2'], df[df['t2'] == 1]['t2'], c='yellow', marker='o', label='Move = 1')

# Etiquetas de los ejes
ax.set_xlabel('x1')
ax.set_ylabel('y1')
ax.set_zlabel('t')

# Mostrar labels
ax.legend()

# Mostrar el grfico
plt.show()
```
